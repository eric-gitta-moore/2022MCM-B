```python
# 相关系数热力图

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Consolas'
plt.rcParams['axes.unicode_minus'] = False  #用来正常显示负号

data = pd.read_excel(r'E:\Programing\PycharmProjects\2022MCM-B\Q1\数据\相关系数.xlsx', sheet_name='区域1')
data.index = data.columns
data2 = pd.read_excel(r'E:\Programing\PycharmProjects\2022MCM-B\Q1\数据\相关系数.xlsx', sheet_name='区域2')
data2.index = data2.columns

# 画出热力图
fig1, ax1 = plt.subplots()
sns.set(font_scale=1.3)
sns.heatmap(data, vmax=1, vmin=-1, center=0, annot=True, ax=ax1, cmap=sns.light_palette("#f9b28c", as_cmap=True))
# ax1.set_title('区域1 皮尔逊相关系数热力图')
plt.savefig('区域1 皮尔逊相关系数热力图.svg')
plt.show()

plt.rcParams['font.family'] = 'Consolas'
plt.rcParams['axes.unicode_minus'] = False  #用来正常显示负号
fig2, ax2 = plt.subplots()
sns.set(font_scale=1.3)
sns.heatmap(data2, vmax=1, vmin=-1, center=0, annot=True, ax=ax2, cmap=sns.light_palette("#f9b28c", as_cmap=True))
# ax2.set_title('区域2 皮尔逊相关系数热力图')
plt.savefig('区域2 皮尔逊相关系数热力图.svg')
plt.show()
```
