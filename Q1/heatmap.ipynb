```python
# 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False  #

data = pd.read_excel(r'E:\Programing\PycharmProjects\2022MCM-B\\.xlsx', sheet_name='1')
data.index = data.columns
data2 = pd.read_excel(r'E:\Programing\PycharmProjects\2022MCM-B\\.xlsx', sheet_name='2')
data2.index = data2.columns

# 
fig1, ax1 = plt.subplots(figsize=(14, 12))
sns.heatmap(data, vmax=1, vmin=-1, center=0, annot=True, ax=ax1, cmap=sns.light_palette("#f9b28c", as_cmap=True))
ax1.set_title('1 ')
plt.savefig('1 .svg')
plt.show()

fig2, ax2 = plt.subplots(figsize=(14, 12))
sns.heatmap(data2, vmax=1, vmin=-1, center=0, annot=True, ax=ax2, cmap=sns.light_palette("#f9b28c", as_cmap=True))
ax2.set_title('2 ')
plt.savefig('2 .svg')
plt.show()
```
