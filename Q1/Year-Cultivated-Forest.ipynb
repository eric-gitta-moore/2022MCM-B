```python
import pandas as pd

df = pd.read_excel(r'E:\Programing\PycharmProjects\2022MCM-B\数据\MCMB数据.xlsx')
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(list(df['年']), list(df['耕地面积']), list(df['森林面积']), color='#e18a3b')

ax.set_xlabel('Year', labelpad=6)
ax.set_ylabel('Cultivated', labelpad=10)
ax.set_zlabel('Forest', labelpad=10)
ax.tick_params(axis='z', pad=6)

plt.savefig('Year-Cultivated-Forest.svg')
plt.tight_layout()
plt.show()
```
